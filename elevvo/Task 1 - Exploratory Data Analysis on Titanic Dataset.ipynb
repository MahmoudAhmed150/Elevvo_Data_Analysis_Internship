Task 1 - Exploratory Data Analysis on Titanic Dataset

```python
import pandas as pd
import csv
import os
```

```python
import glob

path = r"C:\Users\Lenovo\Downloads\titanic"
files = glob.glob(os.path.join(path, "*.csv"))

dfs = {}

for file in files:
    name = os.path.splitext(os.path.basename(file))[0]
    dfs[name] = pd.read_csv(file)

print(dfs.keys())
```

```python
dfs['train']
```

```python
dfs['test']
```

```python
titanic = pd.concat([dfs['test'],dfs['train']], axis = 0)
```

```python
titanic
```

```python
titanic['PassengerId'].nunique()
```

```python
titanic.info()
#--age--, --Fare--, --cabin--, embarked, survived
```

```python
titanic.drop(columns=['Cabin'], inplace=True)
```

```python
titanic['Age'] = titanic.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
titanic['Fare'] = titanic.groupby('Pclass')['Fare'].transform(lambda x: x.fillna(x.median()))
```

```python
titanic['Pclass'].nunique()
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
```

```python
titanic['IsAlone'] = (titanic['FamilySize'] == 1).astype(int)
```

```python
titanic.groupby('Sex')
```

```python
titanic
```

```python
train = titanic[titanic['Survived'].notna()]
```

```python
train
```

```python
survive_ratio = train['Survived'].mean() * 100
survive_ratio
```

```python
train.groupby('Sex')['Survived'].mean()*100
```

```python
train.groupby('Pclass')['Survived'].mean()*100
```

```python
train['AgeGroup'] = pd.cut(train['Age'], bins=[0,12,18,35,60,100])
train.groupby('AgeGroup')['Survived'].mean() * 100
```

```python
train.groupby('FamilySize')['Survived'].mean() * 100
```

```python
train['FareGroup'] = pd.qcut(train['Fare'], 4)
train.groupby('FareGroup')['Survived'].mean() * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=train, x='Survived')
plt.title("Survival Count")
plt.show()
```

```python
sns.barplot(data=train, x='Sex', y='Survived')
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(data=train, x='Pclass', y='Survived')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
sns.histplot(data=train, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.boxplot(data=train, x='Survived', y='Fare')
plt.title("Fare vs Survival")
plt.show()
```

```python
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1

sns.barplot(data=train, x='FamilySize', y='Survived')
plt.title("Survival by Family Size")
plt.show()
```

```python
sns.barplot(data=train, x='Embarked', y='Survived')
plt.title("Survival by Embarkation Port")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(train.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
